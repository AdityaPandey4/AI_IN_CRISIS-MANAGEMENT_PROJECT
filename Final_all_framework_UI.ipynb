**transformers**- 🤗 The Hugging Face Transformers Library is a Python library that provides access to various pre-trained models.

**dataset**- 🤗 Datasets is a library for easily accessing and sharing datasets for Audio, Computer Vision, and Natural Language Processing (NLP) tasks.

**trl**- 🤗TRL is a full stack library where we provide a set of tools to train transformer language models with Reinforcement Learning, from the Supervised Fine-tuning step (SFT)

**peft**- 🤗 PEFT (Parameter-Efficient Fine-Tuning) is a library for efficiently adapting large pretrained models to various downstream applications without fine-tuning all of a model's parameters

**accelerate**-🤗 Accelerate is a library that enables the same PyTorch code to be run across any distributed configuration by adding just four lines of code

**bitsandbytes**-The bitsandbytes library is a lightweight Python wrapper around CUDA custom functions, in particular 8-bit optimizers

**auto-gptq**- The AutoGPTQ library enables users to quantize Transformers models using the GPTQ method.

**optimum**- is an extension of Transformers that provides a set of performance optimization tools to train and run models on targeted hardware with maximum efficiency.

**playwright** - Playwright Library provides unified APIs for launching and interacting with browsers and web scrapping

**Langchain** - Langchain is a framework which allow us to use various llm models as well as allow us to chain multiple tasks such as memory retrieval etc

```python
import gradio as gr
import torch
from datasets import load_dataset
from peft import LoraConfig, prepare_model_for_kbit_training, get_peft_model
from transformers import AutoModelForCausalLM, AutoTokenizer, GPTQConfig, TrainingArguments, pipeline, AutoConfig
from trl import SFTTrainer
import os
import requests
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.document_transformers import Html2TextTransformer
from langchain.document_loaders import AsyncChromiumLoader
from langchain.embeddings.huggingface import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain.prompts import PromptTemplate
from langchain.schema.runnable import RunnablePassthrough
from langchain.llms import HuggingFacePipeline
from langchain.chains import LLMChain
import nest_asyncio
import asyncio
import speech_recognition as sr
from pydub import AudioSegment


# This function is going to perform Google search using the Google Custom (Google programmable Search ) Search API 
def search_google_latest(query, cx, api_key):
    url = "https://www.googleapis.com/customsearch/v1"
    params = {
        "q": query,
        "cx": cx,
        "key": api_key,
        "num": 5,
        "sort": "date:r:0"
    }
    try:
        response = requests.get(url, params=params)
        data = response.json()
        if "items" in data:
            return [item["link"] for item in data["items"]] # returning an array with links of the articles
        else:
            return []
    except Exception as e:
        print("An error occurred:", e)
        return []

def search_and_generate_llm(query,situation):
    # Replace 'YOUR_CX' and 'YOUR_API_KEY' with your own values
    cx = "YOUR_CX"
    api_key = "YOUR_API_KEY"

    search_results = search_google_latest(query, cx, api_key)
    if not search_results:
        return "No relevant links found."

    articles = search_results
    loader = AsyncChromiumLoader(articles)
    docs = loader.load()
    html2text = Html2TextTransformer() # converting all the HTML content to text for Rag
    docs_transformed = html2text.transform_documents(docs)
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=2000, chunk_overlap=0, separators=[" ", ",", "\n"]) # chunking the text into smaller parts
    chunked_documents = text_splitter.split_documents(docs_transformed)

    db = FAISS.from_documents(chunked_documents, HuggingFaceEmbeddings(model_name='sentence-transformers/all-mpnet-base-v2'))
    retriever = db.as_retriever() # Created a Vector Database for the documents to be used by Rag

    
     # Load the language model
    tokenizer = AutoTokenizer.from_pretrained("TheBloke/Mistral-7B-Instruct-v0.2-GPTQ")
    config = AutoConfig.from_pretrained("TheBloke/Mistral-7B-Instruct-v0.2-GPTQ")
    config.quantization_config["disable_exllama"] = True

    model = AutoModelForCausalLM.from_pretrained("TheBloke/Mistral-7B-Instruct-v0.2-GPTQ", low_cpu_mem_usage=True, torch_dtype=torch.float16, config=config)
    pipe = pipeline(task="text-generation", model=model, tokenizer=tokenizer, device="cuda:0", max_new_tokens=2096, temperature=0.1, do_sample=True)
    llm = HuggingFacePipeline(pipeline=pipe)
    # Define the prompt template for the LLM such that it can generate curated the response
    prompt_template = """
                <|system|>\n: You are a highly trained AI assistant designed to provide immediate, accurate, and empathetic support to patients in the aftermath of natural disasters. Your role is to offer guidance, answer questions, and provide information that is both relevant and reassuring. You strive to maintain a calm and supportive demeanor, ensuring that patients receive the help they need in a timely manner.

                  Use the following pieces of retrieved context to give detailed answer to the question.

                  Given the critical nature of the situation, you are equipped with comprehensive knowledge of emergency response protocols, first aid, and mental health support. You are also capable of understanding and responding to a wide range of queries, from immediate medical needs to emotional and psychological support.

                  When faced with a patient's query, you will:
                  - Listen carefully to the patient's concerns.
                  - Provide clear, detailed, and accurate information based on the latest medical guidelines and emergency response protocols.
                  - Offer reassurance and support, acknowledging the patient's fears and concerns.
                  - Direct the patient to seek professional medical attention if necessary, providing guidance on how to do so safely and efficiently.

                  In your responses, you will:
                  - Prioritize the patient's safety and well-being.
                  - Use a friendly and understanding tone, ensuring the patient feels supported and not overwhelmed.
                  - Provide detailed practical advice and resources, including contact information for local emergency services and mental health support.
                  - provide a exact amount of water and food that they should consume in order to maintain their health in good condition

                  Remember, your primary goal is to ensure the patient feels heard, understood, and supported, while also providing them with the necessary information and resources to navigate through the crisis.


                  {context}

                  \n<|user|>\n:
                  {question}
                  """
    # Create a prompt object with the template and input variables to pass to Rag chain
    prompt = PromptTemplate(input_variables=["context", "question"], template=prompt_template)
    llm_chain = LLMChain(llm=llm, prompt=prompt)
    rag_chain = {"context": retriever, "question": RunnablePassthrough()} | llm_chain

    result = rag_chain.invoke(f"{situation}")

    return result['text']
input_query = gr.Textbox(lines=1, placeholder="Enter your query", label="query")
input_situation = gr.Textbox(lines=1, placeholder="enter your situation", label="situation")

# Create a Gradio Interface
iface = gr.Interface(
    fn=search_and_generate_llm,
    inputs=[input_query,input_situation],
    outputs="text",
    title="Public Engagement Chatbot",
    description="This is a Public Engagement Chatbot.We are sorry for your situation and pray that the situation get better soon. kindly provide us the category and the discription of your situation"
)
iface.launch(share=True,debug=True)
```
