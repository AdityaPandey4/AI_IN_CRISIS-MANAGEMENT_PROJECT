```python
import gradio as gr
import torch
from datasets import load_dataset, Dataset
from peft import LoraConfig, prepare_model_for_kbit_training, get_peft_model
from transformers import AutoModelForCausalLM, AutoTokenizer, GPTQConfig, TrainingArguments
from trl import SFTTrainer
import os
import torch
from transformers import (
    AutoModelForCausalLM,
    AutoTokenizer,
    BitsAndBytesConfig,
    pipeline
)
from datasets import load_dataset
from peft import LoraConfig, PeftModel

from langchain.text_splitter import CharacterTextSplitter
from langchain.document_transformers import Html2TextTransformer
from langchain.document_loaders import AsyncChromiumLoader

from langchain.embeddings.huggingface import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS

from langchain.prompts import PromptTemplate
from langchain.schema.runnable import RunnablePassthrough
from langchain.llms import HuggingFacePipeline
from langchain.chains import LLMChain

# Function to handle the chatbot input
def chat_function(name, message):
    # Store inputs in variables
    user_name = name
    user_message = message

    # Example response from the chatbot
    response = f"Hello {user_name}, you said: {user_message}"

    return response

# Define the input components: two textboxes for user inputs
input_name = gr.inputs.Textbox(lines=1, placeholder="Enter your name", label="Name")
input_message = gr.inputs.Textbox(lines=2, placeholder="Enter your message", label="Message")

# Define the output component: a textbox for the chatbot's response
output_response = gr.outputs.Textbox(label="Chatbot Response")

# Create the Gradio interface
iface = gr.Interface(
    fn=chat_function,
    inputs=[input_name, input_message],
    outputs=output_response,
    title="Simple Chatbot",
    description="Enter your name and a message to chat with the bot."
)

# Launch the Gradio interface
iface.launch()
```
